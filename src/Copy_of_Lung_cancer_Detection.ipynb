Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection & Analysis

```python
# loading the data from csv file to a Pandas DataFrame
lg_data = pd.read_csv('/content/Lung_Cancer.csv')
```

```python
# printing the first 5 rows of the dataframe
lg_data.head()
```

```python
# number of rows and columns in the dataframe
lg_data.shape
lg_data.dtypes
```

```python
#Check for missing values
print(lg_data.isnull().sum())

#Remove missing values
df = lg_data.dropna()
```

```python
df.columns = df.columns.str.strip()
#Convert '2' to '1' and '1' to '0'
new_columns = ['SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE',
                      'CHRONIC DISEASE', 'FATIGUE', 'ALLERGY', 'WHEEZING',
                      'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH',
                      'SWALLOWING DIFFICULTY', 'CHEST PAIN']

for col in new_columns:
    if col in df.columns:
        df[col] = df[col].map({2: 1, 1: 0})

# Convert 'YES' to '1' and 'NO' to '0' for the 'LUNG_CANCER' column
lg_data['Target'] = lg_data['Target'].map({'YES': 1, 'NO': 0})

lg_data.head()
```

```python
# getting some statistical measures about the data
lg_data.describe()
```

```python
# checking the distribution of Target Variable
lg_data['Target'].value_counts()
```

1  --> Parkinson's Positive

0 --> Healthy


```python
X = lg_data.drop(columns='Target', axis=1)
Y = lg_data['Target']
```

Data Pre-Processing

Separating the features & Target

```python
# separating the data and labels
X = lg_data.drop(columns = 'Target', axis=1)
Y = lg_data['Target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data to training data & Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

Model Training

Support Vector Machine Model

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```

Model Evaluation

Accuracy Score
